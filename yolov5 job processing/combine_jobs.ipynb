```python
from os.path import join, getmtime
from os import listdir,rename,walk
from pathlib import Path
import distutils
import shutil
from collections import defaultdict
```

```python
labelled_path = join(".","data","labelled_jobs","ignore","yolov5-ir")
```

```python
jobs_path = list(map(lambda x: join(labelled_path,x),listdir(labelled_path)))
jobs_path
```

```python
save_path = join(labelled_path,"combined_job_sig_range_ir")
```

```python
Path(save_path).mkdir(parents=True, exist_ok=True)
```

```python
one = jobs_path[0]
```

```python
shutil.copy(join(one,'data.yaml'),join(save_path,'data.yml'))
```

```python
table = defaultdict(set)
```

```python
def write_splits(split_name):
    
    l_label = join(save_path,split_name,"labels")
    i_label = join(save_path,split_name,"images")
    
    Path(l_label).mkdir(parents=True, exist_ok=False)
    Path(i_label).mkdir(parents=True, exist_ok=False)
    
    def copy(unique_id,src_path,des_path,records):
        print(unique_id,src_path)
        count = 0
        for path, directories, files in walk(src_path):
                for f in files:
                    if f not in records:
                        shutil.copyfile(join(path,f), join(des_path,unique_id+'_'+f if unique_id != "combined_j" else f))
                        count+=1
                        records.add(f)
                    
        print(f"number of files {count}")
                    
    for j_path in jobs_path:
        print(f"Started {j_path}")
        label_path = join(j_path,split_name, "labels")
        img_path = join(j_path,split_name, "images")
        
        uid = Path(j_path).stem.replace(" ","")[:10]
        
        copy(uid,label_path,l_label,table[j_path])
        copy(uid,img_path,i_label,table[j_path])
        
        # distutils.dir_util.copy_tree(label_path,l_label)
        # distutils.dir_util.copy_tree(img_path,i_label)   
        print(f"Finished {j_path}")
        print()
        # print(table)
```

```python
write_splits("train")
write_splits("test")
write_splits("valid")
```
